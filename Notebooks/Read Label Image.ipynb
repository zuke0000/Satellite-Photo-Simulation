```python
from matplotlib.image import imread
import numpy as np
image = imread('Labelled High Res.png')
filtered = np.empty_like(image)
empty_value = image[0,0]
```

```python
for i in range(len(image)):
    for j in range(len(image)):
        match = (np.sum(image[i,j] == empty_value)) == 4
        if (match == False):
            filtered[i,j] = [255,255,255,1]
        else:
            filtered[i,j] = [0,0,0,1]
        
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,18))
plt.imshow(filtered)
plt.show()
```
