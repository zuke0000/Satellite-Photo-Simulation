```python
from matplotlib.image import imread
import numpy as np
image = imread('Chicago_Tutorial_GT.png')
filtered = np.empty_like(image)
```

```python
image[0,70]
```

```python
for i in range(len(image)):
    for j in range(len(image)):
        building_pixel = (np.sum(image[i,j]) > 1.5)
        if (building_pixel):
            filtered[i,j] = [1,1,1,1]
        else:
            filtered[i,j] = [0,0,0,1]
        
```

```python
import matplotlib.pyplot as plt
from PIL import Image
fig = plt.figure(figsize=(15,18))
plt.imshow(filtered)
plt.axis('off')
plt.show()
```

```python
import cv2   
import numpy as np

array = np.array(filtered, np.uint8) * 255
cv2.imwrite('Chicago_Tutorial_GT_modified.png', array)
```
